```python
import sys
sys.path.append('/efs_ecco/ECCO/EMU/emu_userinterface_dir/')
import emu_plot_arg_py as ept
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid',{"grid.linestyle": ":"})
fig_path      = '../figures'
```

```python
globals_dict = ept.emu_plot(run_name='/efs_ecco/lenders/north_mask/emu_atrb_m_5_-56.0_-47.0_50.0_50.0_0.0_0.0_1')
```

```python
return_vars_dict = globals_dict.get('return_vars')
```

```python
lhs    = return_vars_dict['variable'][0]
wind   = return_vars_dict['variable'][1]
htflux = return_vars_dict['variable'][2]
fwflux = return_vars_dict['variable'][3]
sflx   = return_vars_dict['variable'][4]
pload  = return_vars_dict['variable'][5]
ic     = return_vars_dict['variable'][6]
time   = return_vars_dict['time_values']
```

```python
c_lhs    = 'k'
c_wind   = '#cc222b'
c_htflux = '#537c78'
c_fwflux = '#7ba591'
c_sflux  = '#f15b4c'
c_pload  = '#faa41b'
c_ic     = '#ffd45b'
```

```python
sv_name = '/north_face_attribution_ts'
sns.set_style('whitegrid',{"grid.linestyle": ":"})
fig     = plt.figure(figsize=(12, 6))

lhs_    = plt.plot(time,return_vars_dict['variable'][0]/1e6, c_lhs, linewidth = 2, label = 'lhs')
wind_   = plt.plot(time,return_vars_dict['variable'][1]/1e6, c_wind, label = 'wind')
htflux_ = plt.plot(time,return_vars_dict['variable'][2]/1e6, c_htflux, label = 'htflux')
fwflux_ = plt.plot(time,return_vars_dict['variable'][3]/1e6, c_fwflux, label = 'fwflux')
sflx_   = plt.plot(time,return_vars_dict['variable'][4]/1e6, c_sflux, label = 'sflx')
pload_  = plt.plot(time,return_vars_dict['variable'][5]/1e6, c_pload, label = 'pload')
ic_     = plt.plot(time,return_vars_dict['variable'][6]/1e6, c_ic, label = 'ic')

xtix = plt.yticks(fontsize=18)
xtix = plt.xticks(fontsize=18)

plt.xlim(time[0],time[-1])
plt.ylim(-0.6,0.6)
plt.legend(fontsize=15)
plt.ylabel('Flux Anomaly [$Sv$]',fontsize = 18)
plt.savefig(fig_path + sv_name + '.png', format='png', bbox_inches="tight",dpi=500)
```
